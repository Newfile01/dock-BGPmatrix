```python
from matrix_bgpsim import RMatrix

# Charger la matrice
rmatrix = RMatrix.load("/home/jovyan/work/benchmark/rmatrix-torch-core-20000.20250101.as-rel2.lz4")

# Liste pour stocker les AS trouvés et leur type
as_list = []

# Vérifier chaque AS de 1 à 17809
for asn in range(1, 17810):
    if rmatrix.has_asn(asn):
        is_core = rmatrix.is_core_asn(asn)
        is_branch = rmatrix.is_branch_asn(asn)
        as_type = "Core" if is_core else "Branch" if is_branch else "Unknown"
        as_list.append({"asn": asn, "type": as_type})

# Afficher le nombre d'AS trouvés
print(f"Nombre d'AS trouvés : {len(as_list)}")

# Afficher les 10 premiers résultats pour vérification
print("\nExemple des 10 premiers AS trouvés :")
for as_entry in as_list[:10]:
    print(f"AS{as_entry['asn']} ({as_entry['type']})")

# Tenter de vérifier un chemin entre deux AS existants
if len(as_list) >= 2:
    source_asn = as_list[0]["asn"]
    dest_asn = as_list[-1]["asn"]
    as_path = rmatrix.get_path(str(source_asn), str(dest_asn))
    print(f"\nChemin BGP entre AS{source_asn} et AS{dest_asn} : {as_path}")
else:
    print("\nPas assez d'AS trouvés pour vérifier un chemin.")
```

```python
from matrix_bgpsim import RMatrix

# Charger une topologie
rmatrix = RMatrix.load("/home/jovyan/work/benchmark/rmatrix-torch-core-20000.20250101.as-rel2.lz4")

# Lancer la simulation
rmatrix.run(max_iter=32, save_next_hop=True, backend="torch", device="cuda:0")

# Vérifier la présence d'un AS
print("AS 1234 est-il présent ?", rmatrix.has_asn("1234"))

# Obtenir le chemin entre deux AS
path = rmatrix.get_path("1234", "5678")
print("Chemin :", path)

# Obtenir l'état de la route
priority, length = rmatrix.get_state("1234", "5678")
print(f"Priorité : {priority}, Longueur : {length}")
```
